## SVM

```python
# prepare dataset
np.random.seed(3846)
n = 40
omega = np.random.randn()
noise = 0.8 * np.random.randn(n)

x = np.random.randn(n, 2) + 0
y = 2 * (omega * x[:,0] + x[:,1] + noise > 0) - 1

plt.plot(np.extract(y>0, x[:,0]), np.extract(y>0, x[:,1]), 'x')
plt.plot(np.extract(y<0, x[:,0]), np.extract(y<0, x[:,1]), 'o')
```

```python
# hyper parameter
lam = 1
alpha_init = np.random.randn(40) + 0.5
K = np.dot(x, x.T) * np.dot(y[:, None], y[None, :])
lip = 1.01 * np.max(np.linalg.eigvals(2 * K))  # 最大固有値
eta = 1/lip
numepoch = 500

# proximal gradient method
def prox(alpha, lb, ub):
    alpha = alpha - (alpha-1)*(alpha > ub) - alpha*(alpha < lb)
    return alpha

def train(numepoch, alpha_init, K, eta, lam):
    whistory = []
    ahistory = []
    vhistory = []
    dvhistory = []
    alpha = alpha_init
    for t in range(numepoch):
        dual_value = - np.einsum('i, ij, j', alpha, K, alpha) / (4*lam) + np.sum(alpha)
        w = np.einsum('i, i, ik', alpha, y, x)
        value = np.sum(np.fmax(0, np.einsum('ij, i, j -> i', x, y, w))) + lam*np.dot(w,w)
        whistory.append(w)
        vhistory.append(value)
        dvhistory.append(dual_value)
        ahistory.append(alpha)
        grad = (np.dot(K, alpha) - 1)  / (2*lam)
        alpha = alpha - eta*grad
        alpha = np.clip(alpha, 0., 1.)  # alpha = prox(alpha, 0, 1)
    ahistory = np.vstack(ahistory)
    whistory = np.vstack(whistory)
    return ahistory, whistory, vhistory, dvhistory

ahistory, whistory, vhistory, dvhistory = train(numepoch, alpha_init, K, eta, lam)
plt.plot(vhistory, label='hinge loss + regularization')
plt.plot(dvhistory, label='dual Lagurange function')
plt.legend()
plt.show()
```

```python
%matplotlib inline
        
w = whistory[-1]
X1, X2 = np.mgrid[np.min(x[:,0]):np.max(x[:,0]):0.01, np.min(x[:,1]):np.max(x[:,1]):0.01]
X = np.stack([X1, X2], axis=2)
fValue = np.einsum('ijk, k -> ij', X, w)
plt.contour(X1, X2, fValue, [-1, 0, 1])

outp = np.dot(x,w)
plt.plot(np.extract((y>0) & (outp>1), x[:,0]), np.extract((y>0)  & (outp>1), x[:,1]), 'bx', markersize=4)
plt.plot(np.extract((y>0) & (outp<=1), x[:,0]), np.extract((y>0)  & (outp<=1), x[:,1]), 'bx', markersize=9)
plt.plot(np.extract((y<0) & (outp>=-1), x[:,0]), np.extract((y<0)  & (outp>=-1), x[:,1]), 'ro', markersize=9)
plt.plot(np.extract((y<0) & (outp<-1), x[:,0]), np.extract((y<0)  & (outp<-1), x[:,1]), 'ro', markersize=4)
```
